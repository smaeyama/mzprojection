***(i) Import the function from the module "mzprojection"***

```python
import sys
sys.path.append("../src/")
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from mzprojection import mzprojection_multivariate_discrete_time, split_long_time_series, calc_correlation

help(mzprojection_multivariate_discrete_time)
```

***(ii) Prepare ensemble of time series data***  
Here, we split a long time-series data of statistically steady state into a number of samples of short-time data.

```python
#= Read sample data =
ds = xr.open_dataset('../sample_data/sample_multivariate_L96xy.nc')
t_raw    = np.array(ds.t)      # Time t
u_raw    = np.array(ds.u).T    # Variable of interest u(t)
f_raw    = np.array(ds.f).T    # Analyzed data f(t)
print("t_raw[:].shape=",t_raw.shape)
print("u_raw[:,nu].shape=",u_raw.shape)
print("f_raw[:,nf].shape=",f_raw.shape,"# nu/=nf is available.")

fig = plt.figure(figsize=(14,2.5))
ax = fig.add_subplot(111)
ax.set_xlabel("Time $t_n$")
ax.set_ylabel("Long time-series $u_i(t_n), f_i(t_n)$")
ax.plot(t_raw,f_raw[:,0],label="$f_i(t_n)$")
ax.plot(t_raw,u_raw[:,0],label="$u_i(t_n)$")
ax.set_xlim(0,None)
plt.legend()
plt.show()
```

```python
#= Split a long time series data into samples of short-time data =
ista    = 1000                # Start time step number for sampling   
nperiod = 400                 # Time step length of a short-time sample                 
nshift  = 10                  # Length of time shift while sampling
t   =split_long_time_series(t_raw,ista=ista,nperiod=nperiod,nshift=nshift)
u   =split_long_time_series(u_raw,ista=ista,nperiod=nperiod,nshift=nshift)
f   =split_long_time_series(f_raw,ista=ista,nperiod=nperiod,nshift=nshift)

fig = plt.figure(figsize=(14,2.5))
samples = [0,250,500]
for i in range(len(samples)):
    ax = fig.add_subplot(int("14{:}".format(i+1)))
    isample = samples[i]
    ax.set_xlabel(r"Time $t_n$")
    ax.set_ylabel(r"Short-time samples $u_i(t_n), f_i(t_n)$")
    ax.plot(t[isample,:],f[isample,:,0],label=r"$f_i(t_n)$")
    ax.plot(t[isample,:],u[isample,:,0],label=r"$u_i(t_n)$")
    ax.legend()
fig.tight_layout()
plt.show()
```

***(iii) Apply the Moti-Zwanzig projection operator method***  
The projection of $\boldsymbol{f}(t)$ on $\boldsymbol{u}(t)$ is calculated as,  
    $\boldsymbol{f}(t_n)=\Omega \cdot \boldsymbol{u}(t_n)+\boldsymbol{s}(t_n)+\boldsymbol{r}(t_n)$,  
    $\boldsymbol{s}(t)=-\sum_{l=1}^n \Gamma(t_l) \cdot \boldsymbol{u}(t_{n-l})$.  
The Markov coefficient $\Omega$, the memory function $\Gamma(t)$ and the uncorrelated term $\boldsymbol{r}(t)$ are obtained as outputs.  
(Some correlations, e.g., $\langle \boldsymbol{r}(t) \boldsymbol{u} \rangle$ are also obtained to check the result.)

```python
omega, memoryf, s, r = mzprojection_multivariate_discrete_time(u, f, flag_terms=True, flag_debug=True)
t_cor = t_raw[0:memoryf.shape[0]]
```

***(iv-a) Check results: plot $\Omega$ and $\Gamma(t)$***  

```python
print("Markov coefficient matrix Omega = ", omega[0,0])

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel("Correlation time $t_n$")
ax.set_ylabel("Memory function $\Gamma(t_n)$")
ax.plot(t_cor,memoryf[:,0,0],label="$\Gamma(t_n)$")
ax.set_xlim(0,None)
plt.legend()
plt.show()
```

***(iv-b) Check results: plot time evolution of $\boldsymbol{f}(t_n)=\Omega \cdot \boldsymbol{u}(t_n)+\boldsymbol{s}(t_n)+\boldsymbol{r}(t_n)$***  

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel(r"Time $t_n$")
ax.set_ylabel(r"$f_i(t_n)=\Omega_{ij}u_j(t_n)+s_i(t_n)+r_i(t_n)$")
isample=200
ax.plot(t[isample,:],f[isample,:,0],label=r"$f_i(t_n)$")
ax.plot(t[isample,:],np.dot(u[isample,:,:],omega.T)[:,0],label=r"$\Omega_{ij}u_j(t_n)$")
ax.plot(t[isample,:],s[isample,:,0],label=r"$s_i(t_n)$")
ax.plot(t[isample,:],r[isample,:,0],label=r"$r_i(t_n)$")
plt.legend()
plt.show()
```

***(iv-c) Check results: Extended generalized fluctuation-dissipation theorem $\Gamma(t) = \langle \boldsymbol{r}(t_n) \boldsymbol{u}(t_1) \rangle \cdot \langle \boldsymbol{u} \boldsymbol{u}(t_1) \rangle^{-1}$***  

```python
u0 = u[:,0]
f0 = f[:,0]
r0 = r[:,0]
uu = calc_correlation(u,u0)
fu = calc_correlation(f,u0)
ru = calc_correlation(r,u0)
ru1 = calc_correlation(r,u[:,1])
uu0_inv = np.linalg.inv(uu[0,:,:])
u0u1_inv = np.linalg.inv(uu[1]).T

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel("Correlation time $t_n$")
ax.set_ylabel(r"$\Gamma_{ij}(t_n) = \langle r_i(t) u_k(t_1) \rangle \langle u u(t_1) \rangle^{-1}_{k,j}$")
ax.plot(t_cor,memoryf[:,0,0],"-",label="$\Gamma_{ij}(t_n)$")
ax.plot(t_cor[1:],np.dot(ru1,u0u1_inv)[1:,0,0],"--",label=r"$\langle r_i(t_{n}) u_k(t_1) \rangle \langle u u(t_1) \rangle^{-1}_{k,j}$",dashes=[3,3],linewidth=3)
plt.legend()
plt.xlim(0,1.5)
plt.show()
```

***(iv-d) Check results: $\langle \boldsymbol{r}(t) \boldsymbol{u} \rangle = 0$***  
Namely, the uncorrelated term $\boldsymbol{r}(t)$ indeed extracts the part of $\boldsymbol{f}(t)$ uncorrelated with $\boldsymbol{u}$. It is checked by $\langle \boldsymbol{r}(t) \boldsymbol{u} \rangle \ll \langle \boldsymbol{f}(t) \boldsymbol{u} \rangle$.

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel("Correlation time $t_n$")
ax.set_ylabel(r"Correlation $\langle r_i(t_n) u_j \rangle = 0$")
ax.plot(t_cor,ru[:,0,0],label=r"$\langle r_i(t_n) u_j \rangle$")
ax.plot(t_cor,fu[:,0,0],label=r"$\langle f_i(t_n) u_j \rangle$")
plt.legend()
plt.show()
```
